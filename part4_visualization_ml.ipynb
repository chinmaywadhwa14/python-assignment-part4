```python
# Counting how many students passed and failed
df['passed'].value_counts()
```

```python
# Calculating average scores for each subject based on pass/fail

subject_cols = ['math', 'science', 'english', 'history', 'pe']

print("Average scores of students who passed:\n")
print(df[df['passed'] == 1][subject_cols].mean())

print("\nAverage scores of students who failed:\n")
print(df[df['passed'] == 0][subject_cols].mean())
```

```python
# Finding the student with highest average score

# Creating a new column for average score
df['avg_score'] = df[['math','science','english','history','pe']].mean(axis=1)

# Getting the top student
top_student = df.loc[df['avg_score'].idxmax()]

print("Top performing student:\n")
top_student
```
